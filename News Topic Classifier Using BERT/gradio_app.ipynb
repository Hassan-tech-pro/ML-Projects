```python
import gradio as gr
from transformers import pipeline

# 1. Point to the folder where you saved the model in the previous task
model_path = "./saved_bert_news_model"

# 2. Initialize the pipeline
# This automatically loads the config, tokenizer, and weights
try:
    classifier = pipeline(
        "text-classification", 
        model=model_path, 
        tokenizer=model_path
    )
    print("Model loaded successfully!")
except Exception as e:
    print(f"Error loading model: {e}. Make sure the path is correct!")

# 3. Mapping the numeric labels back to categories
labels = {
    "LABEL_0": "World",
    "LABEL_1": "Sports",
    "LABEL_2": "Business",
    "LABEL_3": "Sci/Tech"
}

def classify_news(text):
    if not text.strip():
        return "Please enter some text."
    
    # Get prediction
    prediction = classifier(text)[0]
    
    label = labels.get(prediction['label'], "Unknown")
    confidence = round(prediction['score'] * 100, 2)
    
    return f"Prediction: {label} ({confidence}%)"

# 4. Launch Gradio
demo = gr.Interface(
    fn=classify_news,
    inputs=gr.Textbox(label="Headline/Article Text", placeholder="Enter news here..."),
    outputs=gr.Textbox(label="Analysis"),
    title="AG News Classifier",
    theme="soft",
    examples = [
        ["The championship game went into overtime after a dramatic last-minute goal."],
        ["New quantum computing breakthroughs could revolutionize data encryption."],
        ["The central bank raised interest rates today to combat rising inflation."]
    ]
# Add examples=examples inside gr.Interface()
)

demo.launch()
```
